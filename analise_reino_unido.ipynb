# Análise por coluna

Taxa de desemprego

```python
desemprego = df[['Description', 'Unemployment rate']].dropna()
```

```python
desemprego.shape
```

```python
desemprego.rename(columns={'Description': 'Ano', 'Unemployment rate': 'Desemprego'}, inplace=True)
```

```python
desemprego
```

```python
desemprego['Ano'] = desemprego['Ano'].astype('int64')
```

```python
desemprego = desemprego.set_index(keys=['Ano']).sort_values(by='Ano')
```

```python
desemprego.head()
```

```python
desemprego['Desemprego'] = desemprego['Desemprego'].str.replace(',', '.').astype(float)
```

<h3>Os anos com as menores taxas de desemprego</h3>

```python
desemprego.groupby(desemprego.index)['Desemprego'].min().sort_values(ascending=True).head(5)
```

<h3>Os anos com a maior taxa de desemprego do UK</h3>

```python
desemprego.groupby(desemprego.index)['Desemprego'].max().sort_values(ascending=False).head(5)
```

<h3>Media do desemprego</h3>

```python
media = desemprego['Desemprego'].mean()
print(f'Media do desemprego {media}')
```

<h3>Linha do tempo do desemprego</h3>

```python
sns.lineplot(data=desemprego, x=desemprego.index, y='Desemprego')
plt.title('Taxa Do desemprego No Reino Unido')
```
